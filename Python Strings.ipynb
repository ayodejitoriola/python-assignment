## Python Basics

### Variables
- A variable is a name that points to a value stored in the computer memory.

#### Variable Types
- Integers: These are whole numbers like -2, 0, 5, e.t.c.
- Floats: These are fractional numbers like, 2.35, -1.7, e.t.c.

- NB: Python variables are dynamically typed. This means that, the type of a variable is interpreted during runtime, and you
do not need to specify it, unlike other programming languages like C++, and Java.

### Naming Convention
- Reserved words such as sum, class, yield, cannot be used as variable names.
- Variable names cannot start with a number.
- Whitespace character is not allowed in the naming of a variable.
